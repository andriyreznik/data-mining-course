```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
```

Параметри розподілу й об'єм вибірки

```python
a, b = 2, 3
n = 650
```

Генеруємо випадкову вибірку за бета-розподілом

```python
x = beta.rvs(a, b, size=n)
```

Будуємо гістограму частот

```python
plt.hist(x, density=True, bins=20, histtype='stepfilled', alpha=0.5)
plt.ylabel('Density')
plt.xlabel('x-values')
plt.title('Beta Destribution')
plt.text(0.7, 1.5, r'$\xi$' + '~B(2,3)')
```

Будуємо графіки розсіювання вибірки, щільність розподілу, теоретичну та емпіричну функції розподілу

```python
plt.figure(1, figsize=(15,3))
plt.subplot(131)
plt.scatter(range(n), x)
plt.ylabel('Probability')
plt.xlabel('N values')

plt.subplot(132)
sorted_x = np.sort(x)
k = beta.pdf(sorted_x, a, b)
plt.plot(sorted_x, k, 'r-')
plt.ylabel('Density')
plt.xlabel('x-values')

plt.subplot(133)
f = beta.cdf(sorted_x, a, b)
plt.xlabel('')
plt.ylabel('Percentile')
plt.xlabel('x-values')
plt.plot(sorted_x, f, label='CDF')
plt.plot(sorted_x, np.linspace(0, 1, n), label='ECDF')

plt.suptitle('Beta Destribution')
plt.legend()
```

Обчислюємо основні характеристики:

*   Стандартне відхилення (Standard deviation);
*   Дисперсія (Variance);
*   Мат. сподівання (Expected value, mean);


```python
print(np.std(x), np.var(x), np.mean(x))

```
